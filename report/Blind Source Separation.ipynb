# Blind Source Separation for Upmixing Audio
Blind Source Separation in audio aims at recovering each source signal from a set of audio mixtures of the original sources. This is a common problem in signal processing and remains an active area of research

Consider a cocktail party where there are several politicians having animated conversations with lobbyists seeking their vote. In the background is a jazz band playing some slick tunes

## Mathematical Models
Suppose we have $M$ sensors $[x_1[n], \ldots, x_M[n]]^\intercal$ and $N$ sources $[s_1[n], \ldots, s_N[n]]^\intercal$ the signal on each source can be modeled as (in the case of audio, our sensors are microphones and $\mathbf{x}_i = [x_1[i], \ldots, x_T[n]]$ is the samples in an audio file representing the time signal).

$$ x_i[n] = \sum_{j=1}^N \pi_{ij}[n] $$

where $\pi_{ij}[n]$ is the $sound\ image$ of source $j$ on sensor $i$. We see that our sensors are exposed to a sum of sound images created by each of the sources. There are several models for how exactly this sound image is generated, which tradeoff simplicity of the model with fidelity to physical truth.

### Instantaneous/Linear Model
In this model, the mixture on a sensor is formed by a linear combination of the source signals. Thus we have that

$$ x_i[n] = \sum_{j=1}^N a_{ij}s_j[n] $$

where $a_{ij}$ is a scalar for all $i$, $j$. This is a simple linear model and can be rewritten as the system of equations

$$\begin{eqnarray}
    x_1[n] &=& a_{11}s_1[n] + a_{12}s_2[n] + \ldots + a_{1N}s_N[n] \\
    x_2[n] &=& a_{21}s_1[n] + a_{22}s_2[n] + \ldots + a_{2N}s_N[n] \\
    \vdots &=& \vdots \\
    x_M[n] &=& a_{M1}s_1[n] + a_{M2}s_2[n] + \ldots + a_{MN}s_N[n]
\end{eqnarray}$$

This system can be rewritten in matrix form as 
$$ \mathbf{X} = \mathbf{AS} $$

Though this model is unrealistic (not accounting for delays to different microphones, etc.), its power comes from its simplicity.
